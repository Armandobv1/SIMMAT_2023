1. Ajuste estos datos a polinomios de grado 1 hasta 3 (los datos en $x$ corresponden a los años de 2014 a 2040, sin embargo son valores muy grandes para elevarlos a una potencia, use una representación para estos años, por ejemplo $2014 \rightarrow 0$ y $2040\rightarrow 26$).

2. Muestre las tres curvas ajustadas junto a los datos de los costos en un solo gráfico

```python
import numpy as np
import matplotlib.pyplot as plt


years = np.arange(0, 27)
data = np.array([10.051101,  9.259127,  9.444813,  9.557723,  9.977324, 10.298326,
                 10.447217, 10.550506, 10.590784, 10.657418, 10.743085, 10.831965,
                 10.988435, 11.153055, 11.313326, 11.557165, 11.860136, 12.122511,
                 12.356448, 12.611391, 12.814569, 13.111372, 13.357134, 13.559012,
                 13.85861 , 14.135924, 14.436149])

c1 = np.polyfit(years, data, 1)
c2 = np.polyfit(years, data, 2)
c3 = np.polyfit(years, data, 3)

poly_1 = np.poly1d(c1)
poly_2 = np.poly1d(c2)
poly_3 = np.poly1d(c3)

años = np.linspace(0, 26, 100)

fit_1 = poly_1(años)
fit_2 = poly_2(años)
fit_3 = poly_3(años)

# Graficar
plt.figure(figsize=(10, 6))
plt.scatter(years, data, label='Datos reales', marker='o', color='blue')
plt.plot(años, fit_1, label='Polinomio de grado 1', linestyle='--', color='red')
plt.plot(años, fit_2, label='Polinomio de grado 2', linestyle='--', color='green')
plt.plot(años, fit_3, label='Polinomio de grado 3', linestyle='--', color='orange')

plt.xlabel('Años (desde 2014 a 2040)')
plt.ylabel('Costos')
plt.title('Ajuste de datos a polinomios de grado 1 a 3')
plt.legend()
plt.grid(True)
plt.show()
```
